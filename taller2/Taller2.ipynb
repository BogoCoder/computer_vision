#### 1(b). Cree una funcin para realizar un ajuste de contraste de tal forma que mapee los valores de intensidad  de la imagen de entrada a nuevos valores de intensidad en la imagen de salida.  La funcin debe usarse de la siguiente forma `im_out=image_adjust(im_in, range_in= (low_in, high_in), range_out=(low_out, high_out), gamma)`. La funcin mapea los valores entre low_in y high_in a valores entre low_out y hight out, de tal forma que los valores por debajo de low_in y por encima de high_out son cortados.  La funcin debe aceptar cualquier tipo de imagen (en escala de grises), y el resto de parmetros (que no es la imagen), debe estar entre 0 y 1 (siendo cero el valor mnimo de intensidad y 1 correspondiente al valor mximo del formato, por ejemplo en uint8, 1 corresponde a 255).  El parmetro gamma especifica la forma de la curva que mapea la intensidad de los valores de la imagen de entrada a la de salida. La siguiente figura aclara como debe ser la forma que produce dicha funcin:
![](images/ajustarimagen.png)

Sugerencia: Para convertir un rango de imgenes de (a,b) a (c, d) se puede utilizar la ecuacin de una recta : $$y = ((d-c) / (b-a)) * (x-a) + c$$ Sin embargo, es mejor usar una ecuacin ms generalizada:
$$y = ((x-a) / (b-a)) ^ \gamma * (d-c) + c$$ Si $\gamma$ es igual a 1, entonces se usa la ecuacin lineal.
Cuando gamma no es igual a 1, la transformacin no es lineal. Es recomendable usar `skimage.util.img_as_float`.

```python
def image_adjust(im_in, range_in = (0, 1), range_out =(0, 1), gamma = 1):
    """Documente y complete la funcion"""
    I = skimage.util.img_as_float(im_in)

    a = range_in[0]
    b = range_in[1]
    c = range_out[0]
    d = range_out[1]
    
    img_out = (((I-a)/(b-a))**gamma)*(d-c) + c 
    
    img_out = np.minimum(img_out, d)
    img_out = np.maximum(img_out, c)
    
    return img_out
```

```python
new_breast = image_adjust(breast, range_in= (0, 1), range_out=(0, 1), gamma=1)

plt.imshow(new_breast, cmap = 'gray')
```

#### 2(a). Realice una funcin de convolucin 
Cree una funcin de convolucin (sin reflejar el filtro, es decir una correlacin). Recuerde que la convolucin consiste en mover el filtro e ir ponderando una vecindad de pixeles (dependiendo del tamao del filtro) y sustituyendo el pixel central por el valor ponderado:
![](images/convolution.png)

La funcin debe tener la forma de `convolucion_imagen(imagen, filtro, preservar)` y debe rellenar los bordes con ceros para mantener el mismo tamao de la imagen original. Para imgenes en escala de grises es la clsica convolucin en la que se pondera los valores. Sin embargo, si es una imagen con mltiples canales (o componentes) se debe particularizar los casos mediante el parmetro de `preservar`:

- Si `preservar` es `True` se debe producir una imagen con los mismos canales que la imagen de entrada. Esto significa que cada canal en la imagen debe ser separadamente filtrado por el mismo kernel de filtro.

- Si `preservar` es `False` se debe retornar una imagen de escala de grises (un canal), lo cul es producido mediante la aplicacin del filtro a cada canal, y sumando los canales resultantes.

Recuerde que el kernel debe tener slo un canal (y debe ser verificado). 

La idea (como en el primer ejercicio) es crear una funcin desde "cero", haciendo uso de las facultades de skimage para representar imagenes como arrays de numpy (no mucho ms alla...).


```python
# Solo impar
def convolucion_imagen(imagen, filtro, preservar):
    rowf, colf = filtro.shape
    
    if len(imagen.shape) == 2:
        row, col = imagen.shape
        new = imagen.copy()
        
        # Rellenar imagen
        new = np.concatenate((np.zeros((int((rowf-1)/2), col)), new), axis = 0)
        new = np.concatenate((new, np.zeros((int((rowf-1)/2), col))), axis = 0)
        new = np.concatenate((np.zeros((row + (rowf-1), int((colf-1)/2))), new), axis = 1)
        new = np.concatenate((new, np.zeros((row + (rowf-1), int((colf-1)/2)))), axis = 1)
        
        # Aplicamos filtro
        row, col = new.shape
        for i in range(row-rowf):
            for j in range(col-colf):
                new[int(i + (rowf-1)/2),int(j+ (colf-1)/2)] = np.sum(filtro * new[i:i+rowf, j:j+colf])
                
    else:
        row, col, chan = imagen.shape  
        new = np.ones((row,col))
        for ch in chan:
            for i in range(row-rowf):
                for j in range(col-colf):
                    new = filtro * imagen[i:i+rowf, j:j+colf]

    return new        
```

```python
filtro = np.ones((5,5))/25

conv_breast = convolucion_imagen(breast, filtro, True)
```
